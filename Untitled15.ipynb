```python
import numpy as np

# Example Data
# X -> Experience (Years), y -> Salary
X = np.array([1, 2, 3, 4, 5])
y = np.array([25000, 30000, 35000, 40000, 45000])

# Step 1: Compute means
mean_x = np.mean(X)
mean_y = np.mean(y)

# Step 2: Compute slope (m)
numerator = np.sum((X - mean_x) * (y - mean_y))
denominator = np.sum((X - mean_x) ** 2)
m = numerator / denominator

# Step 3: Compute intercept (b)
b = mean_y - (m * mean_x)

# Step 4: Prediction function
def predict(x):
    return m * x + b

# Test
pred = predict(6)
print(f"Predicted salary for 6 years experience: {pred}")
print(f"Equation of line: y = {m:.2f}x + {b:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
X = np.array([[1], [2], [3], [4], [5]])  # must be 2D for sklearn
y = np.array([25000, 30000, 35000, 40000, 45000])

# Step 1: Create model
model = LinearRegression()

# Step 2: Train (fit) the model
model.fit(X, y)

# Step 3: Check learned parameters
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)

# Step 4: Predict for new value
pred = model.predict([[6]])
print("Predicted salary for 6 years experience:", pred[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# LINEAR REGRESSION USING GRADIENT DESCENT
# ===============================

# Generate some sample data
# y = 4x + 7 + noise
np.random.seed(42)
X = 2 * np.random.rand(100, 1)   # 100 random X values between 0 and 2
y = 4 * X + 7 + np.random.randn(100, 1)  # Add some noise to make it realistic

# Visualize the data
plt.scatter(X, y, color='blue', label="Training Data")
plt.title("Sample Data for Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Add a column of 1s to X for the bias term (intercept)
# This converts X from (100,1) -> (100,2)
X_b = np.c_[np.ones((100, 1)), X]

# Initialize random weights (theta = [bias, slope])
theta = np.random.randn(2, 1)

# Set learning parameters
learning_rate = 0.1
iterations = 1000
m = len(X_b)  # number of samples

# Gradient Descent Algorithm
for iteration in range(iterations):
    # Compute predictions
    y_pred = X_b.dot(theta)

    # Compute the gradient (partial derivatives)
    gradients = 2/m * X_b.T.dot(y_pred - y)

    # Update theta using the gradient
    theta = theta - learning_rate * gradients

# Print the final parameters (should be close to [7, 4])
print("Learned parameters (theta):", theta.ravel())

# Predict values using the trained model
X_new = np.array([[0], [2]])          # New data points
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)

# Visualize the regression line
plt.scatter(X, y, color='blue', label="Training Data")
plt.plot(X_new, y_predict, color='red', label="Regression Line")
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# ===============================
# EXPLANATION OF KEY STEPS
# ===============================
# 1️⃣ We create sample data that roughly follows y = 4x + 7.
# 2️⃣ We append a column of ones to X to represent the bias term (intercept).
# 3️⃣ Initialize parameters (theta) randomly.
# 4️⃣ In each iteration:
#       - Compute predictions y_pred = X_b.dot(theta)
#       - Compute the gradient (the direction of steepest ascent)
#       - Update theta in the opposite direction of the gradient
# 5️⃣ After many iterations, theta converges close to the true values.
# 6️⃣ Finally, we plot the regression line to visualize how well our model fits the data.
```
